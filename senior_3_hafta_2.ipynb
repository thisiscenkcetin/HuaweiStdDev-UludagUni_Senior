### 2. LoRA Rank ve Alpha Analizi (Orta Seviye)

LoRA konfigürasyonundaki hiperparametrelerin model üzerindeki etkisini inceleyin
- `LoraConfig` nesnesini kullanarak; `r` (rank) değerini 4, 16 ve 64 olarak değiştirin
- Her bir `r` değeri için `model.print_trainable_parameters()` fonksiyonunu çalıştırarak toplam eğitilebilir parametre sayısını kaydedin
- Rank artışının eğitilebilir parametre sayısı ve dolaylı olarak bellek kullanımı üzerindeki etkisini bir tablo veya metin hücresi olarak raporlayın

```python
from peft import LoraConfig, TaskType, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer

model_adi = "sshleifer/tiny-gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_adi)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

sonuc_listesi = []

for rank in [4, 16, 64]:
    model = AutoModelForCausalLM.from_pretrained(model_adi)
    config = LoraConfig(
        r=rank,
        lora_alpha=rank * 2,
        lora_dropout=0.05,
        bias="none",
        task_type=TaskType.CAUSAL_LM,
        target_modules=["c_attn"],
    )
    model = get_peft_model(model, config)
    model.print_trainable_parameters()
    egitilebilir = sum(param.numel() for param in model.parameters() if param.requires_grad)
    toplam = sum(param.numel() for param in model.parameters())
    sonuc_listesi.append((rank, egitilebilir, toplam))

print("\nLoRA rank özeti")
print("rank | eğitilebilir parametre | toplam parametre")
for rank, egitilebilir, toplam in sonuc_listesi:
    print(f"{rank:>4} | {egitilebilir:>22} | {toplam:>15}")
```

### 3. QLoRA ile Model Eğitimi (İleri Seviye)

Google Colab (T4 GPU) kullanarak küçük bir modeli (Örn: `TinyLlama/TinyLlama-1.1B-Chat-v1.0` veya `microsoft/phi-2`) 4-bit quantization (QLoRA) ile eğitin
- `SFTTrainer` kullanarak eğitimi başlatın
- Eğitimi en az 50-100 adım (steps) sürdürün
- Eğitim sonunda "Training Loss" (Eğitim Kaybı) değerinin düştüğünü gösteren bir grafik veya log çıktısı paylaşın
- Eğitilen LoRA adaptörünü yerel bir dizine kaydedin

*Kütüphaneler: transformers, peft, trl, bitsandbytes*

```python
from pathlib import Path

import matplotlib.pyplot as plt
import torch
from datasets import load_dataset
from peft import LoraConfig, prepare_model_for_kbit_training, get_peft_model
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, TrainingArguments

if not torch.cuda.is_available():
    print("GPU bulunamadı. QLoRA eğitimi yerelde atlandı: bu hücre Colab T4 için hazır")
else:
    from trl import SFTTrainer

    model_adi = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
    veri_yolu = Path("kafe_asistani_veri_seti.jsonl")
    veri_seti = load_dataset("json", data_files=str(veri_yolu), split="train")

    def metin_hazirla(ornek):
        return f"### Talep:\n{ornek['instruction']}\n\n### Yanıt:\n{ornek['output']}"

    veri_seti = veri_seti.map(lambda ornek: {"text": metin_hazirla(ornek)})

    tokenizer = AutoTokenizer.from_pretrained(model_adi)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    bnb_config = BitsAndBytesConfig(
        load_in_4bit=True,
        bnb_4bit_quant_type="nf4",
        bnb_4bit_compute_dtype=torch.float16,
        bnb_4bit_use_double_quant=False,
    )

    model = AutoModelForCausalLM.from_pretrained(
        model_adi,
        quantization_config=bnb_config,
        device_map="auto",
    )
    model = prepare_model_for_kbit_training(model)

    lora_config = LoraConfig(
        r=8,
        lora_alpha=16,
        lora_dropout=0.05,
        bias="none",
        task_type="CAUSAL_LM",
        target_modules=["q_proj", "v_proj"],
    )
    model = get_peft_model(model, lora_config)

    args = TrainingArguments(
        output_dir="qlora_cikti",
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        max_steps=60,
        learning_rate=2e-4,
        logging_steps=10,
        save_steps=30,
        fp16=True,
        optim="paged_adamw_32bit",
        report_to=[],
    )

    trainer = SFTTrainer(
        model=model,
        tokenizer=tokenizer,
        train_dataset=veri_seti,
        dataset_text_field="text",
        max_seq_length=256,
        args=args,
    )

    sonuc = trainer.train()
    model.save_pretrained("qlora_adaptoru")
    tokenizer.save_pretrained("qlora_adaptoru")

    kayitlar = [item["loss"] for item in trainer.state.log_history if "loss" in item]
    print("Eğitim tamamlandı")
    print(sonuc)
    print("Kayıp değerleri:", kayitlar)

    if kayitlar:
        plt.plot(kayitlar, marker="o")
        plt.title("Eğitim Kaybı")
        plt.xlabel("Log adımı")
        plt.ylabel("Kayıp")
        plt.grid(True)
        plt.show()
```

### 4. "Before & After" Çıkarım Testi (Orta Seviye)

Fine-tuning işleminin modelin "aklındaki" bilgiyi nasıl değiştirdiğini test edin
- Eğitmediğiniz "Base Model" (Ana Model) ile kendi eğittiğiniz "Fine-Tuned Model"i (Adaptör takılmış haliyle) karşılaştırın
- Her iki modele de aynı 3 farklı promptu verin
- Çıktıları yan yana koyarak; modelin sizin veri setinizdeki formata, dile veya bilgiye ne kadar uyum sağladığını analiz edin

```python
from pathlib import Path

import torch
from peft import PeftModel
from transformers import AutoModelForCausalLM, AutoTokenizer

model_adi = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
adaptor_klasoru = Path("qlora_adaptoru")
promptlar = [
    "Müşteriye bugün yoğunluk olduğunu kısa ve nazik şekilde söyle",
    "Bir kahve önerisi ver",
    "Rezervasyon isteyen müşteriye kısa cevap yaz",
]

if not adaptor_klasoru.exists():
    print("Önce eğitim hücresini çalıştırmak gerekiyor")
else:
    device_map = "auto" if torch.cuda.is_available() else {"": "cpu"}
    dtype = torch.float16 if torch.cuda.is_available() else torch.float32

    tokenizer = AutoTokenizer.from_pretrained(model_adi)
    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token

    def cevap_uret(model, metin):
        girdiler = tokenizer(metin, return_tensors="pt").to(model.device)
        cikti = model.generate(
            **girdiler,
            max_new_tokens=60,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
        )
        return tokenizer.decode(cikti[0], skip_special_tokens=True)

    base_model = AutoModelForCausalLM.from_pretrained(model_adi, torch_dtype=dtype, device_map=device_map)
    ince_ayarli_taban = AutoModelForCausalLM.from_pretrained(model_adi, torch_dtype=dtype, device_map=device_map)
    ince_ayarli_model = PeftModel.from_pretrained(ince_ayarli_taban, str(adaptor_klasoru))

    for prompt in promptlar:
        base_yanit = cevap_uret(base_model, prompt)
        fine_yanit = cevap_uret(ince_ayarli_model, prompt)
        print(f"\nİstem: {prompt}")
        print(f"Temel model: {base_yanit}")
        print(f"İnce ayarlı model: {fine_yanit}")
```

### 5. Chat Template Uygulaması (İleri Seviye)

Veri setinizi modelin orijinal eğitimindeki "Chat Template" (Sohbet Şablonu) yapısına uygun hale getirin
- `tokenizer.apply_chat_template` fonksiyonunu kullanarak ham metinlerinizi modelin (Örn: Llama-3 veya Mistral) anladığı formatta (special tokens dahil) hazırlayın
- Neden ham metin eğitmek yerine bu şablonları kullanmamız gerektiğini, "senior" bir bakış açısıyla (modelin önceden öğrenmiş olduğu dizilimleri korumak vb.) açıklayın

```python
import os
from transformers import AutoTokenizer

os.environ["HF_HUB_DISABLE_SYMLINKS_WARNING"] = "1"

model_adi = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(model_adi)

mesajlar = [
    {"role": "system", "content": "Sen kısa ve net cevap veren bir kafe asistanısın"},
    {"role": "user", "content": "Bugün için hafif bir kahve öner"},
]

sablonlu_metin = tokenizer.apply_chat_template(mesajlar, tokenize=False, add_generation_prompt=True)
print(sablonlu_metin)
print("\nŞablon kullanınca modelin alıştığı sıra korunuyor, bu yüzden sonuç daha tutarlı oluyor")
```
