Bu bölümde, modern büyük dil modellerinin (LLM) matematiksel ve mimari temellerini derinlemesine inceliyoruz. Bu yapı taşları, PyTorch veya TensorFlow kullanarak production-grade model eğitmek isteyen ileri seviye geliştiricilerin master etmesi gereken core konulardır.

# 1️⃣ Transformer Nedir?
Transformer, 2017'de "Attention is All You Need" makalesiyle tanıtılan, günümüzde GPT, LLaMA, PaLM, Claude, Mistral gibi tüm büyük modellerin temelini oluşturan bir mimaridir.

## Transformer'ın temel farkı:

  - Sıralı işlem yerine paralel işlem yapabilmesi

  - Daha uzun bağlamları anlayabilmesi

  - Bağlam içi ilişkileri doğrudan Attention ile modellemesi


## Transformer iki ana bloktan oluşur:

- Encoder: Girdi anlamlandırılır (BERT gibi modeller)

- Decoder: Metin üretilir (GPT gibi modeller)

Günümüzde LLM'ler genellikle decoder-only yapıdadır.

# 2️⃣ Attention Mekanizması
Attention, modelin bir cümledeki kelimelerin birbirleriyle olan ilişkisini matematiksel olarak öğrenmesini sağlar.

Basit örnek:
"Araba hızla virajı döndü çünkü lastikleri çok iyiydi."
Burada lastikleri kelimesi, araba ile doğrudan ilişkilidir. Attention matrisi, bu ilişkileri sayısal olarak öğrenir.

Nasıl çalışır:
Her token için üç vektör oluşturulur:

  - Q (Query): Bu token ne arıyor?

  - K (Key): Bu token ne sunuyor?

  - V (Value): Bu token'ın taşıdığı bilgi

Model bu üç vektörü kullanarak hangi kelimelerin birbirine daha çok dikkat etmesi gerektiğini öğrenir.

Multi-Head Attention: Bu ilişkiyi farklı uzaylarda paralel olarak hesaplar. Böylece model aynı anda farklı türde ilişkileri (gramer, anlam, bağlam) yakalayabilir.

# 3️⃣ Tokenizasyon
LLM'ler kelimeyi değil, daha küçük birim olan token'ı işler.

Örnek:
"transformation" → ["trans", "form", "ation"]

"Türkiye'nin" → ["Türk", "iye", "'", "nin"]

## Neden token kullanırız?

  - Kelime çeşitliliğini azaltmak

  - Nadir kelimeler için esnek parçalama yapmak

  - Model performansını artırmak

## Popüler türleri:

- BPE (Byte Pair Encoding): GPT serisi kullanır
- WordPiece: BERT kullanır
- SentencePiece: LLaMA, T5 kullanır

Token sayısı çıktı uzunluğunu, maliyeti ve model kapasitesini direkt etkiler.

#4️⃣ Embedding
Tokenlar sayı değildir ve anlamı yoktur. Bu yüzden her token, çok boyutlu bir vektöre (örneğin 4096 boyut) dönüştürülür.

Örnek:

"king"  → [0.24, -0.11, 0.89, ...]

"queen" → [0.22, -0.10, 0.91, ...]

Bu vektörler:

- Kelimenin anlamını

- Bağlam içindeki rolünü

- İlişkilerini temsil eder


Positional Encoding: Transformer'da kelimelerin sırasını anlamak için her token'a pozisyon bilgisi eklenir. Bu sayede model "kedi köpeği kovaladı" ile "köpek kediyi kovaladı" arasındaki farkı anlayabilir.

Embedding katmanı eğitime dahil edilir; yani model çalıştıkça vektörler "zekileşir".

# 5️⃣ Inference (Çıkarım Süreci)

Inference, eğitim bitmiş bir modelin yeni bir metin üretmesine denir.

Süreç:

  - Prompt → tokenize edilir

  - Tokenlar embedding'e çevrilir

  - Transformer blokları çalışır

  - Çıkış tokenı için olasılık dağılımı hesaplanır

  - Sampling/decoding yapılır (greedy, top-k, top-p)

  - Yeni token üretilir

  - Bu yeni token girişe eklenir, tekrar 1. adıma dönülür

Decoding stratejileri:

- Greedy: Her seferinde en olası kelimeyi seç (hızlı ama tekrarlayıcı)

- Top-k: En olası k kelime arasından rastgele seç

- Top-p (Nucleus): Belirli bir olasılık eşiğine ulaşana kadar kelimeleri topla

- Temperature: Modelin ne kadar "yaratıcı" olacağını kontrol eder

Bu döngü saniyede yüzlerce kez yapılabilir ve böylece model akıcı metinler üretir.
