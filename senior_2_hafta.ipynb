# 1️⃣ Hugging Face Ekosistemi Nedir?
Hugging Face, yapay zeka dünyasının "GitHub"ı konumundadır. İçerisinde binlerce hazır model, veri seti ve bu modelleri çalıştırmak için optimize edilmiş kütüphaneler barındırır.

## Temel Bileşenler:
- **Model Hub:** LLaMA, BERT, Mistral gibi modellerin ağırlıklarının (weights) paylaşıldığı depo.
- **Transformers Kütüphanesi:** Modelleri tek satırla indirmeyi, konfigüre etmeyi ve çalıştırmayı sağlar.
- **Datasets Kütüphanesi:** Çok büyük verileri RAM'i şişirmeden (streaming) işleyebilen yapıdır.
- **Tokenizers Kütüphanesi:** 1. haftada gördüğümüz BPE, WordPiece gibi tekniklerin çok hızlı (Rust tabanlı) uygulamalarıdır.

# 2️⃣ Model Mimarileri: Encoder vs. Decoder
Transformer mimarisi 1. haftada gördüğümüz üzere iki ana bloktan oluşur. Modern NLP'de modeller görevlerine göre bu bloklardan birine odaklanır:

- **Encoder-Only (BERT, RoBERTa):** Metni her iki yönden (bi-directional) okur. Metnin "anlamını" çıkarmada en iyisidir. 
  - *Görevler:* Sınıflandırma, Duygu Analizi, NER (Varlık Tanıma).
- **Decoder-Only (GPT Serisi, LLaMA, Mistral):** Metni soldan sağa (uni-directional) okur. Bir sonraki token'ı tahmin etmek (üretim) için optimize edilmiştir.
  - *Görevler:* Metin Üretimi, Chatbotlar, Kod Yazımı.
- **Encoder-Decoder (T5, BART):** Bir girdi dizisini alıp bambaşka bir çıktı dizisine dönüştürür.
  - *Görevler:* Çeviri, Özetleme (Summarization).

# 3️⃣ Pipeline API: En Hızlı Çıkarım (Inference)
Hugging Face `pipeline` fonksiyonu; tokenizasyon, model yükleme ve çıktı üretme adımlarını tek bir çatı altında birleştirir. Karmaşık modelleri saniyeler içinde test etmemizi sağlar.

````python
# Gerekli kütüphaneyi yükleyelim
# !pip install transformers datasets

from transformers import pipeline

# 1. Örnek: Türkçe Duygu Analizi (Sentiment Analysis)
# Model Hub'dan popüler bir Türkçe BERT modelini çağırıyoruz
classifier = pipeline("sentiment-analysis", model="dbmdz/bert-base-turkish-cased-sentiment")

print("Duygu Analizi Testi:")
print(classifier("Bu eğitim kampı gerçekten çok öğretici ve vizyoner!"))
print(classifier("Maalesef projenin bu kısmında çok fazla hata ile karşılaştık."))

print("\n" + "="*30 + "\n")

# 2. Örnek: Metin Üretimi (Text Generation)
generator = pipeline("text-generation", model="redmond-ai/Gemma-2b-it-Turkish-FineTuning") # Küçük bir örnek model
prompt = "Yapay zeka gelecekte tıp dünyasını nasıl etkileyecek?"
output = generator(prompt, max_length=100, num_return_sequences=1)

print("Metin Üretimi:")
print(output[0]['generated_text'])
````

# 5️⃣ Veri Seti Yönetimi (Hugging Face Datasets)
Senior bir geliştirici için veriyi sadece yüklemek yetmez, onu verimli bir şekilde işlemek (mapping), filtrelemek ve belleği optimize kullanmak gerekir. `datasets` kütüphanesi, veriyi RAM yerine disk üzerinden (memory-mapping) okuyarak TB'larca veriyi işlemenize olanak tanır.

- **Streaming Mode:** Veri setinin tamamını indirmeden, ihtiyaç duydukça internetten parça parça çekme.
- **Map & Filter:** Veri kümesi üzerinde paralel işlemler yapma.

````python
from datasets import load_dataset

# 1. Veri Setini Yükleme (Örnek: Türkçe haber başlıkları veya benzeri)
dataset = load_dataset("savasy/ttc4900", split="train") # Örnek bir Türkçe veri seti

# 2. Keşif
print(f"Veri Seti Yapısı: {dataset}")
print(f"İlk Örnek: {dataset[0]}")

# 3. Veriyi İşleme (Mapping)
# Her metnin uzunluğunu hesaplayan basit bir fonksiyon
def add_length(example):
    example["text_length"] = len(example["text"])
    return example

dataset = dataset.map(add_length)

# 4. Filtreleme
# Sadece 100 karakterden uzun olanları tutalım
long_texts = dataset.filter(lambda x: x["text_length"] > 100)
print(f"Filtrelenmiş Veri Sayısı: {len(long_texts)}")
````

# 6️⃣ Tokenizasyon: Teori'den Uygulamaya
1. haftada tokenizasyonun mantığını görmüştük. Hugging Face'de her modelin kendine has bir tokenizer'ı vardır ve bu tokenizer sadece metni bölmez; **Padding, Truncation ve Special Tokens** (CLS, SEP, PAD) gibi işlemleri de halleder.

- **Padding:** Farklı uzunluktaki cümleleri aynı boyuta getirmek için eklenen "boş" tokenlar.
- **Truncation:** Modelin maksimum girdi limitini (örn: 512) aşan kısımları kesme.
- **Attention Mask:** Modelin hangi tokenlara bakması gerektiğini, hangilerinin padding olduğunu söyler.

````python
from transformers import AutoTokenizer

model_checkpoint = "dbmdz/bert-base-turkish-cased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

sentences = [
    "Bugün hava çok güzel.",
    "Yapay zeka gelişim kampı senior seviye müfredatı devam ediyor."
]

# Uygulamalı Tokenizasyon
inputs = tokenizer(sentences, padding=True, truncation=True, return_tensors="pt")

print("Input IDs (Sayısal Temsil):")
print(inputs["input_ids"])

print("\nAttention Mask (Modelin odaklanacağı yerler):")
print(inputs["attention_mask"])

# Tokenları geri çözme (Decoding)
print("\nSayıların Metin Karşılığı:")
print(tokenizer.convert_ids_to_tokens(inputs["input_ids"][0]))
````
